```python
# Portfolio: Practical Tasks 4 - Pandas Data Exploration and Cleaning

import pandas as pd
import numpy as np

print("--- TASK 18: Load Data from CSV, Excel, JSON ---")
# Loading the Titanic dataset directly from a public URL
titanic_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(titanic_url)

# Saving to JSON and Excel locally in Colab, then loading them back to demonstrate the skill
df.to_json("titanic_data.json")
df.to_excel("titanic_data.xlsx", index=False)

df_json = pd.read_json("titanic_data.json")
print("Successfully loaded CSV, JSON, and Excel files.\n")

print("--- TASK 19: Explore Data ---")
print(f"Dataset Shape: {df.shape} (Rows, Columns)\n")
print("First 3 rows (head):\n", df.head(3), "\n")
print("Last 2 rows (tail):\n", df.tail(2), "\n")
print("Basic Info (columns and data types):")
df.info() # Prints directly
print("\nStatistical Summary (describe):\n", df.describe().round(2), "\n")

print("--- TASK 20: Select Columns & Boolean Indexing ---")
# Selecting specific columns
names_and_fares = df[['Name', 'Fare']]

# Boolean Indexing: Filter for passengers in 1st Class who survived
first_class_survivors = df[(df['Pclass'] == 1) & (df['Survived'] == 1)]
print(f"Number of 1st Class Survivors: {len(first_class_survivors)}\n")

print("--- TASK 21: Sort Data by Multiple Columns ---")
# Sorting by Ticket Class (Ascending) and then by Fare (Descending)
sorted_df = df.sort_values(by=['Pclass', 'Fare'], ascending=[True, False])
print("Top 3 rows after sorting by Class and Fare:\n", sorted_df[['Name', 'Pclass', 'Fare']].head(3), "\n")

print("--- TASK 22: Handle Missing Data ---")
print("Missing values before cleaning:\n", df.isnull().sum()[df.isnull().sum() > 0], "\n")

# 1. Fill missing 'Age' values with the median age
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)

# 2. Drop the 'Cabin' column entirely since it has too many missing values
df = df.dropna(axis=1, thresh=500) # Drops columns with less than 500 non-NA values

# 3. Drop remaining rows with any missing values (e.g., in 'Embarked')
df = df.dropna()
print("Missing values after cleaning:\n", df.isnull().sum(), "\n")

print("--- TASK 23: Add, Remove, and Rename Columns ---")
# Add: Create a 'FamilySize' column (Siblings/Spouses + Parents/Children + 1 for themselves)
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Rename: Change 'Pclass' to 'PassengerClass'
df = df.rename(columns={'Pclass': 'PassengerClass', 'Fare': 'TicketPrice'})

# Remove: Drop the 'SibSp' and 'Parch' columns since we combined them
df = df.drop(columns=['SibSp', 'Parch'])

print("Final dataset columns after modifications:")
print(list(df.columns))
print("\nSample of modified data:\n", df[['Name', 'FamilySize', 'PassengerClass', 'TicketPrice']].head(3))
```
