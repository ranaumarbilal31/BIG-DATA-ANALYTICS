```python
# Portfolio: Task 2 - Python Data & File Handling

import csv
import json
import os

print("--- 1. PYTHON DATA STRUCTURES ---")
# Using tuples for fixed subject names, sets for unique departments
subjects = ("Programming", "Calculus", "Physics")
departments = {"BS-CS", "BS-SE", "BS-IT", "BS-CS"} # Notice duplicate BS-CS will be removed
print(f"Subjects (Tuple): {subjects}")
print(f"Unique Departments (Set): {departments}\n")

# Using a list of dictionaries for our dataset
students_data = [
    {"id": 101, "name": "Saad Ali", "grades": [85, 90, 88]},
    {"id": 102, "name": "Ayesha Khan", "grades": [78, 82, 80]},
    {"id": 103, "name": "Ahmed Raza", "grades": [92, "Absent", 95]}, # Intentional bad data for error handling
    {"id": 104, "name": "Fatima Noor", "grades": [88, 91, 89]}
]

print("--- 2 & 5. FUNCTIONS & ERROR HANDLING ---")
def calculate_average(grades_list):
    """Calculates average grade, safely ignoring non-numeric values."""
    total = 0
    count = 0
    for grade in grades_list:
        try:
            # This will trigger a TypeError if the grade is "Absent" (a string)
            total += float(grade)
            count += 1
        except (ValueError, TypeError):
            print(f"  [Warning] Handled invalid grade entry: '{grade}'")
        except Exception as e:
            print(f"  [Error] An unexpected error occurred: {e}")

    # Prevent ZeroDivisionError
    if count == 0:
        return 0
    return round(total / count, 2)

# Test the function on our data
for student in students_data:
    avg = calculate_average(student['grades'])
    print(f"Student {student['name']} Average: {avg}")


print("\n--- 3. READ AND WRITE CSV & JSON FILES ---")
csv_filename = "student_grades.csv"
json_filename = "student_data.json"

# Write to CSV
print(f"Writing data to {csv_filename}...")
with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["ID", "Name", "Grades"]) # Header
    for student in students_data:
        writer.writerow([student['id'], student['name'], student['grades']])

# Write to JSON
print(f"Writing data to {json_filename}...")
with open(json_filename, 'w') as file:
    json.dump(students_data, file, indent=4)

# Read from text/CSV file
print(f"\nReading back from {csv_filename}:")
with open(csv_filename, mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)


print("\n--- 4. LIST COMPREHENSIONS FOR DATA TRANSFORMATION ---")
# Let's extract just the valid numeric grades from the first student and apply a 5-point curve
student_1_grades = students_data[0]['grades']
print(f"Original Grades: {student_1_grades}")

# List comprehension: [expression for item in iterable if condition]
curved_grades = [grade + 5 for grade in student_1_grades if isinstance(grade, (int, float))]
print(f"Curved Grades (+5): {curved_grades}")
```
